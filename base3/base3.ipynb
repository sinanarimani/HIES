```python
dic1 = {
    'Address': 'HHID',
    'DYCOL01': 'Tenure',
    'DYCOL02': 'Othertenure',
    'DYCOL03': 'Room',
    'DYCOL04': 'Area',
    'DYCOL05': 'Skeleton',
    'DYCOL06': 'MainMaterial',
    'DYCOL07': 'PrivateVehicle',
    'DYCOL08': 'Motorcycle',
    'DYCOL09': 'Bike',
    'DYCOL10': 'Radio',
    'DYCOL11': 'Cassette',
    'DYCOL12': 'TVbw',
    'DYCOL13': 'TV',
    'DYCOL14': 'VideoPlayer',
    'DYCOL15': 'Computer',
    'DYCOL16': 'Cellphone',
    'DYCOL17': 'Freezer',
    'DYCOL18': 'Refrigerator',
    'DYCOL19': 'Freez_refrig',
    'DYCOL20': 'Oven',
    'DYCOL21': 'Vacuum',
    'DYCOL22': 'Washer',
    'DYCOL23': 'Sewing',
    'DYCOL24': 'Fan',
    'DYCOL25': 'Cooler_water_portable',
    'DYCOL26': 'Cooler_gas_portable',
    'DYCOL27': 'Dishwasher',
    'DYCOL28': 'None',
    'DYCOL29': 'Pipewater',
    'DYCOL30': 'Electricity',
    'DYCOL31': 'Pipegas',
    'DYCOL32': 'Phone',
    'DYCOL33': 'Internet',
    'DYCOL34': 'Bathroom',
    'DYCOL35': 'Kitchen',
    'DYCOL36': 'Cooler_water',
    'DYCOL37': 'Centralcooler',
    'DYCOL38': 'Centralheat',
    'DYCOL39': 'Pakage',
    'DYCOL40': 'Cooler_gas',
    'DYCOL41': 'Sewer_network',
    'DYCOL42': 'Cookfuel',
    'DYCOL43': 'Heatfuel',
    'DYCOL44': 'Hotwaterfuel'
}
dic2 = {
    'Address': 'HHID',
    'DYCOL001': 'Housing_type',
    'DYCOL002': 'Housing_type_text',
    'DYCOL01': 'Tenure',
    'DYCOL02': 'Othertenure',
    'DYCOL03': 'Room',
    'DYCOL04': 'Area',
    'DYCOL05': 'Skeleton',
    'DYCOL06': 'MainMaterial',
    'DYCOL07': 'PrivateVehicle',
    'DYCOL08': 'Motorcycle',
    'DYCOL09': 'Bike',
    'DYCOL10': 'Radio',
    'DYCOL11': 'Cassette',
    'DYCOL12': 'TVbw',
    'DYCOL13': 'TV',
    'DYCOL14': 'VideoPlayer',
    'DYCOL15': 'Computer',
    'DYCOL16': 'Cellphone',
    'DYCOL17': 'Freezer',
    'DYCOL18': 'Refrigerator',
    'DYCOL19': 'Freez_refrig',
    'DYCOL20': 'oven',
    'DYCOL21': 'Vacuum',
    'DYCOL22': 'Washer',
    'DYCOL23': 'Sewing',
    'DYCOL24': 'Fan',
    'DYCOL25': 'Cooler_water_portable',
    'DYCOL26': 'Cooler_gas_portable',
    'DYCOL27': 'Dishwasher',
    'DYCOL28': 'Microwave',
    'DYCOL29': 'None',
    'DYCOL30': 'Pipewater',
    'DYCOL31': 'Electricity',
    'DYCOL32': 'Pipegas',
    'DYCOL33': 'Phone',
    'DYCOL34': 'Internet',
    'DYCOL35': 'Bathroom',
    'DYCOL36': 'Kitchen',
    'DYCOL37': 'Cooler_water',
    'DYCOL38': 'Centralcooler',
    'DYCOL39': 'Centralheat',
    'DYCOL40': 'Pakage',
    'DYCOL41': 'Cooler_gas',
    'DYCOL42': 'Sewer_network',
    'DYCOL42_14': 'Elevator',
    'DYCOL43': 'Cookfuel',
    'DYCOL44': 'Heatfuel',
    'DYCOL45': 'Hotwaterfuel'
}
fuel_code_dict = {
    # Cooking (Row 1)
    1: "White oil",
    2: "Diesel",
    3: "LPG",
    4: "Natural gas",
    5: "Electricity",
    6: "Firewood and charcoal",
    7: "Animal fuel",
    8: "Coal",
    9: "Other fuels",
    10: "None",

    # Heating (Row 2)
    11: "White oil",
    12: "Diesel",
    13: "LPG",
    14: "Natural gas",
    15: "Electricity",
    16: "Firewood and charcoal",
    17: "Animal fuel",
    18: "Coal",
    19: "Other fuels",
    20: "None",

    # Water heating (Row 3)
    21: "White oil",
    22: "Diesel",
    23: "LPG",
    24: "Natural gas",
    25: "Electricity",
    26: "Firewood and charcoal",
    27: "Animal fuel",
    28: "Coal",
    29: "Other fuels",
    30: "None"
}
property_status_dict = {
    1: "Owner of land and building",
    2: "Owner of building only",
    3: "Rented",
    4: "Mortgaged",
    5: "In exchange for service",
    6: "Free of charge",
    7: "Other"
}
materials_dict = {
    1: "Brick or Iron",
    2: "Brick and Wood",
    3: "Concrete Block",
    4: "All Brick or Stone and Brick",
    5: "All Wood",
    6: "Clay and Wood",
    7: "Clay and Straw",
    8: "Other"
}
years = list(range(89, 99))+ [99] + list(range(1400, 1403))
years = [1403]
for year in years:
    # Path to your .accdb file
    file_path = fr'C:\Users\Mahsa\Desktop\geo\raw_data\{year}\Outfile-{year}.accdb'
    # Connection string (for 64-bit Access Database Engine)
    conn_str = (
        r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
        f'DBQ={file_path};'
    )
    # Connect
    conn = pyodbc.connect(conn_str)
    # List all table names
    tables = [table.table_name for table in conn.cursor().tables(tableType='TABLE')]
    print(f"Year is :{str(year)}".center(50))
    #---------------------------------------------------------------------------------
    Rdata = pd.read_sql(f"SELECT * FROM R{year}P2", conn)
    Udata = pd.read_sql(f"select * from U{year}P2", conn)
    data = pd.concat([Rdata, Udata], axis=0)
    if year < 91:
        data = data[[col for col in data.columns if col in dic1.keys()]]
        data.rename(columns=dic1, inplace=True)
    else:
        data = data[[col for col in data.columns if col in dic2.keys()]]
        data.rename(columns=dic2, inplace=True)
    data["HHID"] = data["HHID"].astype("str")
    mask = data["HHID"].str.contains(r"\.")
    data.loc[mask, 'HHID'] = data.loc[mask, "HHID"].str.split(".").str[0]
    data = data.loc[data["HHID"].notna()]
    
    data = data.map(lambda x: x.strip() if isinstance(x, str) else x)
    data = data.map(lambda x: pd.NA if x == "" else x)
    text_columns = ["HHID","Othertenure"]
    for col in data.columns:
        if col not in text_columns:
            data[col] = pd.to_numeric(data[col], errors = "coerce")
    data["Cookfuel"] = data["Cookfuel"].map(fuel_code_dict)
    data["Heatfuel"] = data["Heatfuel"].map(fuel_code_dict)
    data["Hotwaterfuel"] = data["Hotwaterfuel"].map(fuel_code_dict)
    data["Tenure"] = data["Tenure"].map(property_status_dict)
    data["Skeleton"] = np.where(data["Skeleton"]==1, "Metal",np.where(data["Skeleton"]==2, "concrete", np.where(data["Skeleton"]==3, "Other", pd.NA)))
    data["MainMaterial"] = data["MainMaterial"].map(materials_dict)
    not_boolean = ['HHID', 'Housing_type', 'Housing_type_text', 'Tenure', 'Room', 'Area',
                   'Skeleton', 'MainMaterial','Cookfuel', 'Heatfuel', 'Hotwaterfuel']
    for col in data.columns:
        if col not in not_boolean:
            data.loc[data[col].isna(),col]=0
    data.to_pickle(fr"P2_{year}")
    print(data.shape)
    print(data.isna().sum())
```
